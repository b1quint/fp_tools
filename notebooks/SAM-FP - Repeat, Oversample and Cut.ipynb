```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = np.arange(36) + 1
s = np.zeros_like(z)
s[0] = 200.
s[10] = 100.

plt.plot(z, s, 'bo', label="Original data.")
plt.show()
```

```python
zzz = np.vstack((z - 36, z, z + 36))
sss = np.vstack((s, s, s))

plt.plot(zzz, sss, "bo", label="Repeated data")
plt.show()
```

```python
# Interpolation as it is implemented
z = np.arange(s.size)
z_interpolation = np.linspace(0, s.size, 3 * s.size)

# Interpolate
new_s = np.interp(z_interpolation, z, s)

plt.plot(z_interpolation, new_s, "r.", label="Implemented interpolation.")
plt.show()
```

```python
x = np.arange(36) + 1
print(x)
print(x.size)
print()

x = np.hstack((x - 36, x, x + 36))
print(x)
print(x.size)
print()
              
x = np.linspace(x.min(), x.max(), 3 * x.size)
print(x)
print(x.size)
print()

x = x[108:-108]
print(x)
print(x.size)
print()
```
